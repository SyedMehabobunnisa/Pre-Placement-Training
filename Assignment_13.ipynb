**Question 1**

Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

**Examples:**  
Input: list1 = 5->2->3->8  
list2 = 1->7->4->5  
Output: New list = 5->7->4->8  
  
Input:list1 = 2->8->9->3  
list2 = 5->3->6->4  
Output: New list = 5->8->9->4  

```python
#Example 1
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def add_node(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    @staticmethod
    def create_new_list(list1, list2):
        if list1 is None or list2 is None:
            return None

        new_list = LinkedList()
        ptr1 = list1.head
        ptr2 = list2.head

        while ptr1 and ptr2:
            if ptr1.value >= ptr2.value:
                new_list.add_node(ptr1.value)
            else:
                new_list.add_node(ptr2.value)
            ptr1 = ptr1.next
            ptr2 = ptr2.next

        return new_list


list1 = LinkedList()
list1.add_node(5)
list1.add_node(2)
list1.add_node(3)
list1.add_node(8)

list2 = LinkedList()
list2.add_node(1)
list2.add_node(7)
list2.add_node(4)
list2.add_node(5)

new_list = LinkedList.create_new_list(list1, list2)

current = new_list.head
while current:
    print(current.value, end=" ")
    current = current.next
```

**Question 3**

Given a linked list of size **N**. The task is to reverse every **k** nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple of *k* then left-out nodes, in the end, should be considered as a group and must be reversed (See Example 2 for clarification).

**Example 1:**  
Input:  
LinkedList: 1->2->2->4->5->6->7->8  
K = 4  
Output:4 2 2 1 8 7 6 5  
Explanation:  
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.

**Example 2:**  
Input:  
LinkedList: 1->2->3->4->5  
K = 3  
Output:3 2 1 5 4  
Explanation:  
The first 3 elements are 1,2,3 are reversed
first and then elements 4,5 are reversed.Hence,
the resultant linked list is 3->2->1->5->4.

```python
#Example 1
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def reverse_k_nodes(head, k):
    current = head
    next = None
    prev = None
    count = 0

    while current is not None and count < k:
        next = current.next
        current.next = prev
        prev = current
        current = next
        count += 1

    if next is not None:
        head.next = reverse_k_nodes(next, k)

    return prev


def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" ")
        temp = temp.next
    print()


head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)


k = 4
head = reverse_k_nodes(head, k)

print("Reversed linked list:")
print_linked_list(head)
```

**Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

**Example:**  
Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3  
Output:   3->2->1->4->5->6->9->8->7->NULL.  

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def reverse_alternate_k_nodes(head, k):
    current = head
    next = None
    prev = None
    count = 0

    while current is not None and count < k:
        next = current.next

        if count % (2 * k) < k:
            current.next = prev
            prev = current
        current = next
        count += 1

    if next is not None:
        head.next = reverse_alternate_k_nodes(next, k)

    return prev


def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" ")
        temp = temp.next
    print()



head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

k = 3
head = reverse_alternate_k_nodes(head, k)

print("Modified linked list:")
print_linked_list(head)
```

**Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

**Examples**:  
Input:   1->2->3->5->2->10, key = 2  
Output:  1->2->3->5->10  

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def delete_last_occurrence(head, key):
    if head is None:
        return head

    prev = None
    curr = head
    last_occurrence = None

    while curr is not None:
        if curr.data == key:
            last_occurrence = curr
        curr = curr.next

    if last_occurrence is None:
        return head

    if last_occurrence == head:
        head = head.next
    else:
        curr = head
        while curr.next != last_occurrence:
            curr = curr.next
        curr.next = last_occurrence.next

    return head


def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" ")
        temp = temp.next
    print()


head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(5)
node5 = Node(2)
node6 = Node(10)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6

print("Original linked list:")
print_linked_list(head)

key = 2
head = delete_last_occurrence(head, key)

print("Modified linked list:")
print_linked_list(head)
```

 **Question 6**

Given two sorted linked lists consisting of **N** and **M** nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**  
Input: a: 5->10->15, b: 2->3->20  

Output: 2->3->5->10->15->20  

Input: a: 1->1, b: 2->4  

Output: 1->1->2->4  

```python
#Example 1
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def merge_sorted_lists(a, b):
    if a is None:
        return b
    if b is None:
        return a

    dummy = Node(0)
    curr = dummy
    prev = dummy

    while a is not None and b is not None:
        if a.data <= b.data:
            curr.next = a
            a = a.next
        else:
            curr.next = b
            b = b.next
        curr = curr.next
        curr.prev = prev
        prev = curr

    if a is not None:
        curr.next = a
        a.prev = prev
    if b is not None:
        curr.next = b
        b.prev = prev

    dummy.next.prev = None

    return dummy.next


def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" ")
        temp = temp.next
    print()


a = Node(1)
a.next = Node(1)

b = Node(2)
b.next = Node(4)

print("First sorted linked list:")
print_linked_list(a)

print("Second sorted linked list:")
print_linked_list(b)

merged = merge_sorted_lists(a, b)

print("Merged sorted linked list:")
print_linked_list(merged)
```

**Question 7**

Given a **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

**Example:**  
Original Linked list 10 8 4 2  
Reversed Linked list 2 4 8 10  

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


def reverse_doubly_linked_list(head):
    if head is None or head.next is None:
        return head

    current = head
    prev = None
    next = None

    while current is not None:
        next = current.next
        current.next = prev
        current.prev = next
        prev = current
        current = next

    head = prev
    return head


def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" ")
        temp = temp.next
    print()


head = Node(10)
node2 = Node(8)
node3 = Node(4)
node4 = Node(2)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3

head = reverse_doubly_linked_list(head)

print("Reversed doubly linked list:")
print_linked_list(head)
```

**Question 8**

Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.

**Example 1:**  
Input:  
LinkedList = 1 <--> 3 <--> 4  
x = 3  
Output:1 3  
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

**Example 2:**
Input:  
LinkedList = 1 <--> 5 <--> 2 <--> 9  
x = 1  
Output:5 2 9  

```python
#Example 1
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


def delete_node(head, position):
    if position == 1:
        if head is None:
            return None
        next_node = head.next
        if next_node is not None:
            next_node.prev = None
        return next_node

    current = head
    count = 1

    while current is not None and count < position:
        current = current.next
        count += 1

    if current is None:
        return head

    current.prev.next = current.next

    if current.next is not None:
        current.next.prev = current.prev

    return head


def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" ")
        temp = temp.next
    print()


head = Node(1)
node2 = Node(3)
node3 = Node(4)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2

position = 3
head = delete_node(head, position)

print("Modified linked list:")
print_linked_list(head)
```
